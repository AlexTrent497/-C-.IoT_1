## Задание 2.2

1. Гипотеза: 

 Если добавить в приложение план-схемы велосипедных дорожек 
(если они присутствуют в населённом пункте) то это дополнительно привлечёт любителей активного отдыха и может увеличить время пользования приложением и число посещений.

2. Что делаем:

- Контрольная версия: оставляем сайт в текущем виде.
- Тестовая версия: добавляем на сайт план-схему велосиподных дорожек

3. На каких пользователях тестируем:

На любителях активного отдыха и на новых пользователях

4. Ключевые метрики для оценки эксперимента:

- Число посещений сайта
- Длитальность нахождения на сайте

5. Ожидаемый эффект:

- Увеличения показателя посящений сайта не 5%
- Увелечения длительности нахождения на сайте на 4%

6. План действий в зависимости от результатов эксперимента:

- Если результат эксперимента будет положительным и прирост значений
ключевых метрик будет не менее ожидаемого, то распространяем изменения на всех
пользователей.
- Если значения ключевых метрик снижаются или имеют незначительный
прирост (менее ожидаемого), то откатываем эксперимент.
